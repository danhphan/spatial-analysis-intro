```python
z = zipfile.ZipFile(atsip_2016_file)
z.printdir()
```

```python
def get_files(zipfile, path):
    file_list = []
    for file in zipfile.namelist():
        if file.startswith(path):
            file_list.append(file)            
    return file_list


file_list = get_files(z, path="2016 Census GCP All Geographies for VIC/SA1/VIC/")
```

```python
for idx, file in enumerate(file_list):
    df = pd.read_csv(z.open(file))
    print(idx, df.shape, file)
```
